```python
import nltk
import re
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Function to clean the text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'\W+', ' ', text)
    # Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Apply the function to the 'text' column
master['text'] = master['text'].apply(preprocess_text)
```

```python
master.head(1)
```

```python
# Assuming 'label' is your target variable
sns.countplot(x='label', data=master)
plt.title('Label Distribution After Text Preprocessing')
plt.show()
```

```python
# Class proportions check after text preprocessing
class_counts = master['label'].value_counts(normalize=True)
print(class_counts)
```
