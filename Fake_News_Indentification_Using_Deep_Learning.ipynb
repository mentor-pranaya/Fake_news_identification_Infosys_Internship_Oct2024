# Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
```

```python
# import main dataset which are we going to use.
ds = pd.read_csv('/content/drive/MyDrive/Fake News Detection Dataset/WithourPreProcessDS.csv')
```

# Data Preprocessing

```python
ds.info()
```

```python
def remove_url(text_data):
  """remove_url takes raw text and removes urls from the text.
     https://stackoverflow.com/questions/11331982/how-to-remove-any-url-within-a-string-in-python/40823105#40823105
     """
  return re.sub(r"http\S+", "", text_data)
for index, str  in zip(ds.index, ds['text']):
  ds.loc[index, 'text'] = remove_url(str)
```

```python
ds.loc[11241, 'text']
```

```python
ds.info()
```

```python
ds['label'].value_counts()
```

```python
ds.loc[:, 'clean_news'] = ds.loc[:, 'text'].str.lower()
ds['clean_news']
```

```python
# Remove stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
ds['clean_news'] = ds['clean_news'].apply(lambda x : " ".join( [ word for word in x.split() if word not in stop ] ))
ds.head()
```

```python
# Remove special Character and Punctuations
ds.loc[:, 'clean_news'] = ds.loc[:, 'clean_news'].str.replace('[^A-Za-z0-9\s]', '',regex=True)
ds.loc[:, 'clean_news'] = ds.loc[:, 'clean_news'].str.replace('\n', '')
ds.loc[:, 'clean_news'] = ds.loc[:, 'clean_news'].str.replace('\s+', ' ',regex=True)
ds['clean_news']
```

```python
# Lemmatization
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def lemmatization(words):
  wl = WordNetLemmatizer()
  lemma_words = []
  for i in words:
  # applying Lemmatization
    lemma_words.append(wl.lemmatize(i))

  return " ".join(lemma_words)

for index in ds.index:
  text = ds.loc[index, 'clean_news']
  # converting text into tokens of word like 'donald trump wish american' to ['donald', 'trump', 'wish', 'american']
  tokens = word_tokenize(text)
  ds.loc[index, 'clean_news'] = lemmatization(tokens)
```
